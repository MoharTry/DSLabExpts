```python
#Task 2 using recursion

def binary(num):
  if num==0:
    return 0
  return num%2+10*binary(num//2)

binary(23)

#Here we are using recursion for calculating the binary form of a given number,using recursion makes the code cleaner and easier for the programmer to calculate the binary form of any number.
```

```python
#Task 3

def fibonacci(n):
  if n==0:
    return 0
  if n==1:
    return 1
  return fibonacci(n-2)+fibonacci(n-1)

fibonacci(20)
```

```python
#Task 2 - Solve a meaningful problem related to recursion which has its use in real life.

# Problem- Given the value of n,r,calculate the number of ways of choosing r things from n things.

def nCr(n,r):
  if r==0 or n==r:
    return 1
  return nCr(n-1,r-1)+nCr(n-1,r)
nCr(5,2)
```
